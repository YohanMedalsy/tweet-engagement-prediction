 # Neural Hawkes Process

```python
import numpy as np

def relu(X):
    res = []
    for x in X:
        if x > 0.0:
            res.append(x)
        else:
            res.append(0.0)
    return res


def scaled_softplus(X, s):
    res = []
    for x in X:
        res.append(s * np.log(1 + np.exp(x / s)))
    return res
```

```python
import matplotlib.pyplot as plt

x = np.linspace(-5, 10, 1500)
s = [0.5, 1, 5, 10]

y_relu = relu(x)
plt.plot(x, y_relu)

for v in s:
    y_ss = scaled_softplus(x, v)
    plt.plot(x, y_ss)
plt.show()
```
